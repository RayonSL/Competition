```python
import random
import cv2
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
from pycocotools.coco import COCO
from pathlib import Path
```

```python
img_prefix = Path('data/dataset/train_images')

coco = COCO('data/dataset/val.json')
```

```python
filename = 'train_restricted_512.jpg'
if filename:
    img_ids = coco.getImgIds()
    for img_info in coco.loadImgs(img_ids):
        if img_info['file_name']==filename:
            img_id = img_info['id']
else:
    img_id = random.choice(coco.getImgIds())
img_info = coco.loadImgs(ids=[img_id])[0]
ann_ids = coco.getAnnIds(imgIds=[img_id])
anns = coco.loadAnns(ids=ann_ids)
img_name = img_info['file_name']

img_filepath = img_prefix/img_name
img = cv2.imread(str(img_filepath))

plt.figure(figsize=(18, 18))
plt.imshow(img)
plt.axis('off')
color_map = [
    [0, 0, 128],
    [100, 149, 237],
    [0, 238, 0],
    [255, 255, 0],
    [255, 48, 48],
]
color_map = (np.array(color_map) / 255).tolist()
coco.showAnns(anns)
print(img_info['file_name'])
```
